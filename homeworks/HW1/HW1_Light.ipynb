```python
def plot_losses(train_losses, train_metrics, val_losses, val_metrics):
    """
    Plot losses and metrics while training
      - train_losses: sequence of train losses
      - train_metrics: sequence of train MSE values
      - val_losses: sequence of validation losses
      - val_metrics: sequence of validation MSE values
    """
    clear_output()
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))
    axs[0].plot(range(1, len(train_losses) + 1), train_losses, label="train")
    axs[0].plot(range(1, len(val_losses) + 1), val_losses, label="val")
    axs[1].plot(range(1, len(train_metrics) + 1), train_metrics, label="train")
    axs[1].plot(range(1, len(val_metrics) + 1), val_metrics, label="val")

    if max(train_losses) / min(train_losses) > 10:
        axs[0].set_yscale("log")

    if max(train_metrics) / min(train_metrics) > 10:
        axs[0].set_yscale("log")

    for ax in axs:
        ax.set_xlabel("epoch")
        ax.legend()

    axs[0].set_ylabel("loss")
    axs[1].set_ylabel("MSE")
    plt.show()

def train_and_validate(
    model,
    optimizer,
    criterion,
    metric,
    train_loader,
    val_loader,
    num_epochs,
    verbose=True,
):
    """
    Train and validate neural network
      - model: neural network to train
      - optimizer: optimizer chained to a model
      - criterion: loss function class
      - metric: function to measure MSE taking neural networks predictions
                 and ground truth labels
      - train_loader: DataLoader with train set
      - val_loader: DataLoader with validation set
      - num_epochs: number of epochs to train
      - verbose: whether to plot metrics during training
    Returns:
      - train_mse: training MSE over the last epoch
      - val_mse: validation MSE after the last epoch
    """
    train_losses, val_losses = [], []
    train_metrics, val_metrics = [], []
    if torch.cuda.is_available():
        device = torch.device("cuda:0")
    else:
        device = torch.device("cpu")
    print(device)

    # Основной цикл по эпохам
    for epoch in range(1, num_epochs + 1):
        # устанавливаем модель в режим обучения
        model.train()
        # Сбрасываем счетчики для текущей эпохи, то что я периодически забываю делать
        running_loss, running_metric = 0, 0
        # Создаем прогрессбар,еле разобрался (если verbose=True)
        pbar = (
            tqdm(train_loader, desc=f"Training {epoch}/{num_epochs}")
            if verbose
            else train_loader
        )

        # цикл по батчам обучающих данных
        for i, (X_batch, y_batch) in enumerate(pbar, 1):
            """
            YOUR CODE HERE (－.－)...zzzZZZzzzZZZ
            Do forward and backward passes
            predictions = ...
            loss = ...
            """
            # переносим батч данных устройство (гпу по возможнсоти)
            X_batch = X_batch.to(device)  # признаки
            y_batch = y_batch.to(device)  # Целевые значения

            # зануление градиента
            optimizer.zero_grad()

            # прямой: получаем предсказания модели
            predictions = model(X_batch)
            # вычисляем функцию потерь между предсказаниями и истинными значениями
            loss = criterion(predictions, y_batch)

            # Бэкпроп, вычисляем гррадиенты
            loss.backward()
            # Обновляем веса
            optimizer.step()

            with torch.no_grad():
                # вычисляем метрику
                metric_value = metric(predictions, y_batch)
                # Если метрика - тензор, преобразуем в число
                if type(metric_value) == torch.Tensor:
                    metric_value = metric_value.item()
                # Накопление потерь и метрик (умножаем на размер батча для усреднения)
                running_loss += loss.item() * X_batch.shape[0]  # loss.item() возвращает значение потерь как число
                running_metric += metric_value * X_batch.shape[0]  # X_batch.shape[0] - размер батча

            # Обновление прогрессбара каждые 100 итераций, хотя самм я такой прогрессбар в жизни не напишу
            # На уроке либо я его сам проморгал, когда объясняли, либо этот момент решили опустить
            if verbose and i % 100 == 0:
                pbar.set_postfix({"loss": loss.item(), "MSE": metric_value})

        # Сохраняем средние потери и метрики за эпоху
        train_losses += [running_loss / len(train_loader.dataset)]  # Средние потери
        train_metrics += [running_metric / len(train_loader.dataset)]  # Средняя метрика

        # Переключаем модель в режим валидации
        model.eval()
        # Сбрасываем счетчики для валидации
        running_loss, running_metric = 0, 0
        # Создаем прогрессбар для валидации
        pbar = (
            tqdm(val_loader, desc=f"Validating {epoch}/{num_epochs}")
            if verbose
            else val_loader
        )

        # Цикл по батчам валидационных данных
        for i, (X_batch, y_batch) in enumerate(pbar, 1):
            # без вычисления градиентов сразу на валидации
            with torch.no_grad():
                """
                YOUR CODE HERE (－.－)...zzzZZZzzzZZZ
                Do evaluation
                predictions = ...
                loss = ...
                """
                # Переносим батч данных на устройство
                X_batch = X_batch.to(device)
                y_batch = y_batch.to(device)
                # Прямой проход для предиктов
                predictions = model(X_batch)
                # Вычисление потерь
                loss = criterion(predictions, y_batch)

                # Вычисление метрики
                metric_value = metric(predictions, y_batch)
                if type(metric_value) == torch.Tensor:
                    metric_value = metric_value.item()
                # Накоп потерь и метрик
                running_loss += loss.item() * X_batch.shape[0]
                running_metric += metric_value * X_batch.shape[0]

            # И снова обновление прогресбара
            if verbose and i % 100 == 0:
                pbar.set_postfix({"loss": loss.item(), "MSE": metric_value})

        # Сохраняем средние потери и метрики за эпоху валидации
        val_losses += [running_loss / len(val_loader.dataset)]
        val_metrics += [running_metric / len(val_loader.dataset)]

        if verbose:
            plot_losses(train_losses, train_metrics, val_losses, val_metrics)

    if verbose:
        print(f"Validation MSE: {val_metrics[-1]:.3f}")

    # Возвращаем метрики за последнюю эпоху
    return train_metrics[-1], val_metrics[-1]
```

$^1$ **Логит** &mdash; выход последнего линейного слоя, может принимать любые вещественные значения. Если применить Softmax к логитам, то получатся вероятности распределения классов.

```python
import torch.optim as optim

# Преобразуем данные в тензоры PyTorch
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

# Нормализуем целевую переменную
unique_years = sorted(set(y))
# Создаём небольшой словарь, в котором напротив упорядоченного множества года будет идти индекс с 0
dict_year_ind = {year: i for i, year in enumerate(unique_years)}
y_train_class = np.array([dict_year_ind[year] for year in y_train])
y_val_class = np.array([dict_year_ind[year] for year in y_val])
y_test_class = np.array([dict_year_ind[year] for year in y_test])


# Создаем тензоры для целевых переменных
y_train_tensor = torch.tensor(y_train_class, dtype=torch.int64)
y_val_tensor = torch.tensor(y_val_class, dtype=torch.int64)
y_test_tensor = torch.tensor(y_test_class, dtype=torch.int64)

# Создаем Dataset и DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)


# Определение архитектуры нейронной сети
class Model(nn.Module):
    def __init__(self, input_size, out_size, num_classes):
        super(Model, self).__init__()
        self.linear_1 = nn.Linear(input_size, out_size)
        self.relu = nn.ReLU()
        self.linear_2 = nn.Linear(out_size, num_classes)

    def forward(self, x):
        x = self.linear_1(x)
        x = self.relu(x)
        x = self.linear_2(x)
        return x

# Параметры модели
input_size = X_train.shape[1] # Кол-во столбцов
out_size = 128
num_classes = len(unique_years)


model = Model(input_size, out_size, num_classes)

# Определяем лосс и оптимизатор
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=1e-2)
```

```python
# Lambda-функция для вычисления MSE в случае классификации
# Берем класс с максимальным логитом и вычисляем MSE между предсказанным и истинным классом
classification_metric = lambda predictions, targets: ((predictions.argmax(dim=1) - targets).float() ** 2).mean()
# (Этот кусок кода с lambda отработал в дипсике, так как не понял, как написать эту часть по заданию. В промте вбил предыдущую часть своего кода, а так же задание и попросил его написать решения, основываясь на предыдущем коде)

# Обучение модели
num_epochs = 4


train_mse, val_mse = train_and_validate(
    model=model,
    optimizer=optimizer,
    criterion=criterion,
    metric=classification_metric,
    train_loader=train_loader,
    val_loader=val_loader,
    num_epochs=num_epochs,
    verbose=True
)
```
